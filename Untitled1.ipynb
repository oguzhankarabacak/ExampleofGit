```python
import pandas as pd
import numpy as np

# Örnek veri seti boyutu
veri_seti_boyutu = 50000

# Rastgele örnek veri oluşturma
np.random.seed(42)  # Tekrarlanabilirlik için seed ayarı
veri1_feature = np.random.randint(1, 11, size=veri_seti_boyutu)
veri2_feature = np.random.randint(1, 11, size=veri_seti_boyutu)


'''
np.random.seed(42)  # Tekrarlanabilirlik için seed ayarı
veri1_feature = np.random.normal(loc=10, scale=2, size=veri_seti_boyutu)
veri2_feature = np.random.normal(loc=12, scale=3, size=veri_seti_boyutu)
'''

# Veri setlerini DataFrame'e dönüştürme
veri_seti1 = pd.DataFrame({"feature_1": veri1_feature})
veri_seti2 = pd.DataFrame({"feature_1": veri2_feature})

# Veri setlerini CSV dosyasına kaydetme
veri_seti1.to_csv("veri_seti1.csv", index=False)
veri_seti2.to_csv("veri_seti2.csv", index=False)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# İki veri setini yükleyin (örnek olarak CSV dosyaları kullanıyorum)
veri_seti1 = pd.read_csv("veri_seti1.csv")
veri_seti2 = pd.read_csv("veri_seti2.csv")

# İlk feature'ları seçin
feature_adı = "feature_1"
veri1_feature = veri_seti1[feature_adı]-2
veri2_feature = veri_seti2[feature_adı]+2

# Seaborn ile dağılım grafiğini çizin
plt.figure(figsize=(10, 6))
sns.histplot(veri1_feature, color="blue", label="Veri Seti 1")
sns.histplot(veri2_feature, color="pink", label="Veri Seti 2")
plt.title("Feature Distribution Comparison")
plt.xlabel("Feature Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
